```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("data/sudoku.jpg",0)

plt.figure(figsize=(5,5)),plt.title("Original"),plt.imshow(img, cmap="gray"),plt.axis("off");
```

### X Gradient

```python
sobelx = cv2.Sobel(src=img, ddepth=cv2.CV_16S, dx=1, dy=0, ksize=5)

plt.figure(figsize=(5,5)),plt.title("SobelX"),plt.imshow(sobelx, cmap="gray"),plt.axis("off");
```

### Y Gradient

```python
sobely = cv2.Sobel(src=img, ddepth=cv2.CV_16S, dx=0, dy=1, ksize=5)

plt.figure(figsize=(5,5)),plt.title("SobelY"),plt.imshow(sobely, cmap="gray"),plt.axis("off");
```

# Laplacian Gradient

```python
laplacian = cv2.Laplacian(src=img, ddepth=cv2.CV_16S, ksize=5)

plt.figure(figsize=(5,5)),plt.title("Laplacian"),plt.imshow(laplacian, cmap="gray"),plt.axis("off");
```
