```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("data/erol.jpg",0)

plt.figure(figsize=(7,7)),plt.title("Original"),plt.imshow(img, cmap="gray"),plt.axis("off");
```

# Erosion

```python
kernel = np.ones((5,5), dtype=np.uint8)

result = cv2.erode(src=img, kernel=kernel, iterations=1)

plt.figure(figsize=(7,7)),plt.title("Erosion"),plt.imshow(result, cmap="gray"),plt.axis("off");
```

# Dilation

```python
result2 = cv2.dilate(src=img, kernel=kernel, iterations=1)

plt.figure(figsize=(7,7)),plt.title("Dilation"),plt.imshow(result2, cmap="gray"),plt.axis("off");
```

# Creating White Noise for Opening

```python
white_noise = np.random.randint(0,2, size=img.shape[:2])
white_noise = white_noise*255

plt.figure(figsize=(5,5)),plt.title("White Noise"),plt.imshow(white_noise, cmap="gray"),plt.axis("off");
```

```python
noise_img = white_noise + img

plt.figure(figsize=(5,5)),plt.title("White Noise + Original Img."),plt.imshow(noise_img, cmap="gray"),plt.axis("off");
```

# Opening

```python
opening = cv2.morphologyEx(src=noise_img.astype(np.float32), op=cv2.MORPH_OPEN, kernel=kernel)

plt.figure(figsize=(7,7)),plt.title("Opening"),plt.imshow(opening, cmap="gray"),plt.axis("off");
```

# Creating Black Noise for Closing 

```python
black_noise = np.random.randint(0,2, size=img.shape[:2])
black_noise = black_noise*(-255)

plt.figure(figsize=(5,5)),plt.title("Black Noise"),plt.imshow(black_noise, cmap="gray"),plt.axis("off");

# filtering for showing img.
black_noise_img = black_noise + img
black_noise_img[black_noise_img <= -245] = 0

plt.figure(figsize=(5,5)),plt.title("Black Noise Img"),plt.imshow(black_noise_img, cmap="gray"),plt.axis("off");
```

# Closing

```python
closing = cv2.morphologyEx(src=noise_img.astype(np.float32), op=cv2.MORPH_CLOSE, kernel=kernel)

plt.figure(figsize=(7,7)),plt.title("Closing"),plt.imshow(closing, cmap="gray"),plt.axis("off");
```

# Morphological Gradient Operation


$ morphological Gradient = dilation - erosion $

```python
gradient = cv2.morphologyEx(src=img, op=cv2.MORPH_GRADIENT, kernel=kernel)

plt.figure(figsize=(7,7)),plt.title("Gradient"),plt.imshow(gradient, cmap="gray"),plt.axis("off");
```
