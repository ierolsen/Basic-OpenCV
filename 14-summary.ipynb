```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("data/hw.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(6,6)),plt.title("Original RGB"), plt.imshow(img_rgb), plt.axis("off");

print(f" img width: {img.shape[0]},\n img height: {img.shape[1]},\n img channel: {img.shape[2]}")
```

# Resize

```python
img_resized = cv2.resize(img, (int(img.shape[0]*4/5), int(img.shape[1]*4/5)))

img_rgb_resized = cv2.cvtColor(img_resized, cv2.COLOR_BGR2RGB)

print(f" resized img width: {img_resized.shape[0]},\n resized img height: {img_resized.shape[1]}")

plt.figure(figsize=(6,6)),plt.title("Resized RGB Image"), plt.imshow(img_rgb_resized);
```

# Text

```python
cv2.putText(img=img, text="HORSE", org=(10,540), fontFace=cv2.FONT_HERSHEY_COMPLEX, fontScale=2, color=(255,0,0), thickness=2)
cv2.putText(img=img, text="DOG", org=(400,90), fontFace=cv2.FONT_HERSHEY_COMPLEX, fontScale=2, color=(0,255,0), thickness=2)
cv2.putText(img=img, text="CAT", org=(605,140), fontFace=cv2.FONT_HERSHEY_COMPLEX, fontScale=2, color=(0,0,255), thickness=2)


plt.figure(figsize=(6,6)),plt.title("Img with Text"), plt.imshow(img);
```

# Shape

```python
cv2.rectangle(img=img, pt1=(10,10), pt2=(270,550), color=(255,0,0), thickness=4)
cv2.rectangle(img=img, pt1=(350,100), pt2=(560,360), color=(0,255,0), thickness=4)
cv2.rectangle(img=img, pt1=(570,145), pt2=(760,370), color=(0,0,255), thickness=4)


plt.figure(figsize=(7,7)),plt.title("Img with rectange"), plt.imshow(img);
```

# Thresholding

```python
img = cv2.imread("data/hw.jpg")

_, img_thres = cv2.threshold(src=img, thresh=50, maxval=255, type=cv2.THRESH_BINARY)

plt.figure(figsize=(6,6)),plt.title("Thresholding Image"), plt.imshow(img_thres, cmap="gray");
```

# Gaussian Blurring

```python
gb = cv2.GaussianBlur(src=img, ksize=(3,3), sigmaX=10)

plt.figure(figsize=(6,6)),plt.title("Gaussian Blur Image"), plt.imshow(gb), plt.axis("off");
```

# Laplacian Blur

```python
laplace = cv2.Laplacian(src=img, ddepth=cv2.CV_16U)

plt.figure(figsize=(6,6)),plt.title("Laplacian Blur Image"), plt.imshow(laplace), plt.axis("off");
```

# Histogram

```python
img_hist = cv2.calcHist(images=[img], channels=[0], mask=None, histSize=[256], ranges=[0,256])

plt.figure(figsize=(6,6)),plt.title("Histogram"), plt.plot(img_hist), plt.axis("off");
```
