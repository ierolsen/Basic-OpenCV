```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("data/img1.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


plt.title("original")
plt.imshow(img, cmap="gray")
plt.axis("off");
```

# thresholding

```python
_, thres_img = cv2.threshold(src=img, thresh=60, maxval=255, type=cv2.THRESH_BINARY)

plt.title("thresholding | type: THRESH_BINARY")
plt.imshow(thres_img, cmap="gray")
plt.axis("off");
```

```python
_, thres_img = cv2.threshold(src=img, thresh=60, maxval=255, type=cv2.THRESH_BINARY_INV)

plt.title("thresholding | type: THRESH_BINARY_INV")
plt.imshow(thres_img, cmap="gray")
plt.axis("off");
```

# Adaptive Thresholding

```python
thres_img2 = cv2.adaptiveThreshold(src=img, 
                                   maxValue=255, 
                                   adaptiveMethod=cv2.ADAPTIVE_THRESH_MEAN_C, 
                                   thresholdType=cv2.THRESH_BINARY, 
                                   blockSize=11, 
                                   C=8)

plt.title("Adaptive Threshold")
plt.imshow(thres_img2, cmap="gray")
plt.axis("off");
```
