```python
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread("data/img1.jpg")
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB) # BGR to RGB

img2 = cv2.imread("data/img2.jpg")
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(img1)

plt.subplot(1,2,2)
plt.imshow(img2);

print(img1.shape)
print(img2.shape)
```

# resize

```python
img1 = cv2.resize(img1, (600,600))
print(img1.shape)

img2 = cv2.resize(img2, (600,600))
print(img2.shape)

plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(img1)

plt.subplot(1,2,2)
plt.imshow(img2);
```

# Blending

$$ blending = alpha * img1 + beta * img2 $$

```python
blended = cv2.addWeighted(src1=img1, alpha=0.5, src2=img2, beta=0.5, gamma=0)

plt.figure(figsize=(10,10))
plt.title("blended | alpha=0.5 | beta=0.5")
plt.imshow(blended);
```
