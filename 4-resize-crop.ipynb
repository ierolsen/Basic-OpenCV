```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("data/lenna.png")

print(f"Size: {img.shape}")
```

```python
plt.title("original")
plt.imshow(img);
```

# Resize

```python
imgResized = cv2.resize(img, (800,800))

print(f"Size: {imgResized.shape}")
```

```python
plt.title("Resized")
plt.imshow(imgResized);
```

# Crop

```python
imgCropped = img[:200,0:300]

plt.title("Croped")
plt.imshow(imgCropped);
```
