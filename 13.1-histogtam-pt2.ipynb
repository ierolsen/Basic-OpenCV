```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("data/goldenGate.jpg")
img_vis = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(7,7)),plt.title("Original"),plt.imshow(img_vis, cmap="gray"),plt.axis("off");

print(img.shape)
```

### Define Mask

```python
mask = np.zeros(img.shape[:2], np.uint8)

plt.figure(),plt.title("Mask"),plt.imshow(mask, cmap="gray"),plt.axis("off");

#filter
mask[1500:2000, 1000:2000] = 255
plt.figure(),plt.title("Mask with filter"),plt.imshow(mask, cmap="gray"),plt.axis("off");
```

### Apply Mask to Image

```python
masked_img = cv2.bitwise_and(src1=img_vis, src2=img_vis, mask=mask)

plt.figure(figsize=(7,7)),plt.title("Masked Image"),plt.imshow(masked_img, cmap="gray"),plt.axis("off");
```

# Histogram

```python
masked_img = cv2.bitwise_and(src1=img, src2=img, mask=mask)

masked_img_hist = cv2.calcHist(images=[img], channels=[0], mask=mask, histSize=[256], ranges=[0,256])
plt.figure(figsize=(7,7)),plt.title("Masked Image for Red"),plt.plot(masked_img_hist),plt.axis("off");

masked_img_hist = cv2.calcHist(images=[img], channels=[1], mask=mask, histSize=[256], ranges=[0,256])
plt.figure(figsize=(7,7)),plt.title("Masked Image for Green"),plt.plot(masked_img_hist),plt.axis("off");

masked_img_hist = cv2.calcHist(images=[img], channels=[2], mask=mask, histSize=[256], ranges=[0,256])
plt.figure(figsize=(7,7)),plt.title("Masked Image for Blue"),plt.plot(masked_img_hist),plt.axis("off");
```
