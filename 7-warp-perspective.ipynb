```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("data/kart.png")

plt.title("original")
plt.imshow(img);
```

```python
width = 400
height = 500
```

```python
# points of original image
p1 = np.float32([[203,1], 
                 [1,472], 
                 [540,150], 
                 [338, 617]])

# last points of original image after transform
p2 = np.float32([[0,0], 
                 [0,height], 
                 [width,0], 
                 [width,height]])

# Perspective Transform Matrix
matrix = cv2.getPerspectiveTransform(p1, p2)
print(f"Matrix: \n{matrix}")

# Warp Perspective
imgOutput = cv2.warpPerspective(src=img, M=matrix, dsize=(width,height))

# Visualize
plt.title("Warp Perspective")
plt.imshow(imgOutput);
```
