```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("data/red_blue.jpg")
img_vis = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(5,5)),plt.title("Original"),plt.imshow(img_vis, cmap="gray"),plt.axis("off");
```

# Histogram

```python
img_hist = cv2.calcHist(images=[img], channels=[0], mask=None, histSize=[256], ranges=[0,256])

print(img_hist.shape)

plt.figure(figsize=(7,7))
plt.plot(img_hist);
```

```python
color = ("b", "g", "r")
plt.figure(figsize=(7,7))

for i,c in enumerate(color):
    hist = cv2.calcHist(images=[img], channels=[i], mask=None, histSize=[256], ranges=[0,256])
    plt.plot(hist,color = c)
```
