```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# img
img = np.zeros((512,512,3), np.uint8) # black img.

plt.title("black")
plt.imshow(img);
```

# Line

```python
cv2.line(img=img, pt1=(100,100), pt2=(200,200), color=(0,255,0), thickness=2)

plt.figure(figsize=(8,8))
plt.title("with line")
plt.imshow(img);
```

# Rectangle

```python
cv2.rectangle(img=img, pt1=(200,200), pt2=(300,300), color=(0,0,255), thickness=2)

plt.figure(figsize=(8,8))
plt.title("with rectange")
plt.imshow(img);
```

## Filled Rectangle

```python
cv2.rectangle(img=img, pt1=(300,200), pt2=(400,100), color=(255,0,0), thickness=cv2.FILLED)

plt.figure(figsize=(8,8))
plt.title("with Filled")
plt.imshow(img);
```

# Circle

```python
cv2.circle(img=img, center=(400,300), radius=100, color=(255,255,255), thickness=2)

plt.figure(figsize=(8,8))
plt.title("with circle")
plt.imshow(img);
```

## Filled Circle

```python
cv2.circle(img=img, center=(100,100), radius=20, color=(150,150,150), thickness=cv2.FILLED)

plt.figure(figsize=(8,8))
plt.title("with filled circle")
plt.imshow(img);
```

# Text

```python
cv2.putText(img=img, text="OPENCV", org=(50,400), fontFace=cv2.FONT_HERSHEY_DUPLEX, fontScale=2, color=(100,200,100), thickness=2)

plt.figure(figsize=(8,8))
plt.title("with Text")
plt.imshow(img);
```
