```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("data/hist_equ.jpg",0)

plt.figure(figsize=(7,7)),plt.title("Original"),plt.imshow(img, cmap="gray"),plt.axis("off");
```

# Histogram Equalization 

#### This method usually increases the global contrast 

```python
img_hist = cv2.calcHist(images=[img], channels=[0], mask=None, histSize=[256], ranges=[0,256])

plt.figure(figsize=(7,7)),plt.plot(img_hist);

cv2.imshow("hist",img_hist)

key = cv2.waitKey(0) & 0xFF

if key == 27:
    cv2.destroyAllWindows()
```

```python
eq_hist = cv2.equalizeHist(src=img)

plt.figure(figsize=(7,7)),plt.plot(img_hist);

cv2.imshow("equalize hist.",eq_hist)

key = cv2.waitKey(0) & 0xFF

if key == 27:
    cv2.destroyAllWindows()
```

### Histogram of Equalizated Image

```python
eq_img_hist = cv2.calcHist(images=[eq_hist], channels=[0], mask=None, histSize=[256], ranges=[0,256])

plt.figure(figsize=(7,7)),plt.plot(eq_img_hist);
```
