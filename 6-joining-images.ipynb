```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("data/lenna.png")

plt.title("original")
plt.imshow(img);
```

# horizontal joining

```python
hor = np.hstack((img,img))

plt.title("horizontal")
plt.imshow(hor);
```

# vertical joining

```python
ver = np.vstack((img,img))

plt.title("vertical")
plt.imshow(ver);
```
